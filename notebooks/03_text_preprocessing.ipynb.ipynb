```python
import pandas as pd
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
data_path = "../data/raw/Combined Data.csv"
df = pd.read_csv(data_path)
df.head()
```

```python
df = df.dropna(subset=['statement'])
df.isnull().sum()
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def clean_text(text):
    # convert to lowercase
    text = text.lower()
    
    # remove urls
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # remove non-alphabetic characters
    text = re.sub(r'[^a-z\s]', '', text)
    
    # tokenize
    words = text.split()
    
    # remove stopwords and lemmatize
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    
    return " ".join(words)
```

```python
df['cleaned_statement'] = df['statement'].apply(clean_text)
```

```python
df[['statement', 'cleaned_statement']].head()
```

```python
df = df[df['cleaned_statement'].str.strip() != ""]
```

```python
output_path = "../data/processed/cleaned_mental_health.csv"
df.to_csv(output_path, index=False)
```
